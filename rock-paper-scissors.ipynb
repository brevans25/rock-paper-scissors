# Rock, Paper, Scissors

A Python program that plays a game of Rock, Paper, Scissors between a user and the computer.

### Importing Libraries

Import the `random` library to generate a random response from computer

```python
import random
```

#### Store options in a tuple

```python
options = ('rock', 'paper','scissors') # Using tuple because options are inmutable. 
print(options)
```

#### Capture user input
- Capture user input;<br>
- Use `.lower()` to standardize input.

```python
playerChoice = input("Rock, Paper or Scissors? Pick one!").lower()
print(playerChoice)
```

### Generate random choice for the computer

The `choice()` method returns a randomly selected element from the specified sequence.

The sequence can be a string, a range, a list, a tuple or any other kind of sequence.

**SYNTAX** 

`random.choice(sequence)` WHERE `sequence` could be a list, a tuple, a range of numbers etc...


```python
computerChoice = random.choice(options)
print(computerChoice)
```

### Game Logic


- if `input()` IS NOT `'rock'`,`'paper'` OR `'scissors'` THEN display error message.

- if `playerChoice` EQUAL `computerChoice` THEN `TIE`;

- if `playerChoice` BEATS `computerChoice` THEN `WIN`;

- if `computerChoice` EQUAL `playerChoice` THEN `LOSS`.

`PRINT` results

```python
if playerChoice not in options:
    print(f'You entered {playerChoice}, which is not a valid response.')

elif playerChoice == computerChoice:
    print(f'You chose: {playerChoice}. The computer chose: {computerChoice}. It\'s a tie!')

elif (playerChoice == 'rock' and computerChoice == 'scissors') \
    or (playerChoice == 'paper' and computerChoice == 'rock') \
    or (playerChoice == 'scissors' and computerChoice == 'paper'):
    print(f'You chose: {playerChoice}. The computer chose: {computerChoice}. {playerChoice.capitalize()} beats {computerChoice.capitalize()} . You won!')
 
else:
    print(f'You chose: {playerChoice}. The computer chose: {computerChoice}. {computerChoice.capitalize()} beats {playerChoice.capitalize()} . Computer won!')

```
